```python
df["Energy_Consumption(MWh)"].fillna(df["Energy_Consumption(MWh)"].mean(),inplace=True)
```

```python
df["Cost(Million$)"].fillna(df["Cost(Million$)"].mean(),inplace=True)
```

```python
#preprocessing
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

# Instantiate the scaler
min_scale = MinMaxScaler()

# Apply scaling
df[["Energy_Consumption(MWh)", "Cost(Million$)"]] = min_scale.fit_transform(df[["Energy_Consumption(MWh)", "Cost(Million$)"]])
```

```python
df.head()
```

```python
import pandas as pd

df = pd.get_dummies(df, columns=['Energy_Source'])
print(df.head())
```

```python
import matplotlib.pyplot as plt


labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 15, 15]
colors = ['#FFD700', 'blue', 'red', '#32CD32']
explode = (0.5, 0.1, 0.1, 0)  


fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, 
       startangle=120, wedgeprops={'linewidth': 1, 'edgecolor': 'black'})


ax.set_title('Programming Language Popularity', fontsize=14, fontweight='bold')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')


sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')


plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.title('Scatter Plot of Total Bill vs. Tip Amount')
plt.grid(True, linestyle='--', alpha=1) 

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')


pivot_table = tips.pivot_table(values='tip', index='day', columns='size', aggfunc='mean')


plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', linewidths=1)

plt.xlabel('Table Size')
plt.ylabel('Day of the Week')
plt.title('Average Tip Amount by Table Size and Day')

plt.show()
```
