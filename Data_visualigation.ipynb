```python
import matplotlib.pyplot as plt

# Sample data for energy consumption over 6 months (in MWh)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Create a bar chart
plt.bar(months, energy_consumption, color='b', label='Energy Consumption')

# Add titles and labels
plt.title('Energy Consumption Over 6 Months', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12, fontweight='bold')
plt.ylabel('Energy Consumption (MWh)', fontsize=12, fontweight='bold')

# Add a legend
plt.legend(loc='upper left', fontsize=10, frameon=True, shadow=True, fancybox=True)

# Add grid lines
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Customize tick parameters
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 4, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [5, 3, 1, 4, 2]
})

# Create the regression plot
sns.regplot(x='A', y='B', data=data)

# Show the plot
plt.show()


'''regression plot to visualize the relationship between columns 'A' and 'B' in the dataset. 
The sns.regplot function fits a linear regression model to the data 
and plots the regression line along with the scatter plot of the data points.'''
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])
z = np.array([2, 3, 4, 5, 6])

# Create a figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the data
scatter = ax.scatter(x, y, z, c='r', marker='o', s=100, alpha=0.8, edgecolors='w', linewidth=1.5)

# Add labels
ax.set_xlabel('X axis', fontsize=12, labelpad=10)
ax.set_ylabel('Y axis', fontsize=12, labelpad=10)
ax.set_zlabel('Z axis', fontsize=12, labelpad=10)

# Add title
ax.set_title('3D Scatter Plot', fontsize=15, pad=20)



# Show the plot
plt.show()
```
