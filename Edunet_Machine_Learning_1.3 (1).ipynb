```python
import pandas as pd
df=pd.read_csv("Salary_Data.csv")
df.head(10)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#Feature identification
X = df[["YearsExperience"]]
y = df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size=0.2,random_state=32)
```

```python
from sklearn.linear_model import LinearRegression  

lr = LinearRegression()

lr.fit(X_train, y_train) 
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
X_test
```

```python
from sklearn.metrics import r2_score  


r2 = r2_score(y_test, y_pred)  

print("R-squared Score:", r2)
```

```python
#Save the model

import joblib as jb
jb.dump(lr,"Sal_predictor.pkl")
```

```python
model = jb.load('Sal_predictor.pkl')
```

```python
import numpy as np
exp = float(input("Enter your experiance to know your salery"))
new_data = np.array([[exp]])
predict=model.predict(new_data)
print(f"Predicted salary on the experiance input {predict[0]:.2f}INR")
```
